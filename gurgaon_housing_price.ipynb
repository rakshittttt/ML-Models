```python
import pandas as pd
from sklearn.ensemble import GradientBoostingRegressor
```

```python
data=pd.read_excel("housing-train.xlsx")
```

```python
data.head()
```

```python
x_train=data.iloc[:,:-2]
```

```python
y_train=data.iloc[:,-2]
```

```python
y_train.values
```

```python
model=GradientBoostingRegressor(
    n_estimators=200,
    learning_rate=0.05,
    max_depth=3,
    random_state=42)
```

```python
model=model.fit(x_train.values,y_train.values)
```

```python
test_data=pd.read_excel("housing-test.xlsx")
```

```python
x_test=test_data.iloc[:,:-2]
```

```python
prediction=model.predict(x_test.values)
```

```python
actual=test_data.iloc[:,-2]
```

```python
from sklearn.metrics import r2_score
```

```python
r2 = r2_score(actual, prediction)
print("R2 score:", r2)
```
